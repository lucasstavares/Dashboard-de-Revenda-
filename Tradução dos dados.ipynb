```python
import numpy as np
import random
import string
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./retail_sales_dataset.csv')
```

```python
df = df.replace(to_replace='Male', value='Masculino', regex=True)
```

```python
df = df.replace(to_replace='Female', value='Feminino', regex=True)
```

```python
df = df.rename(columns={"Transaction ID": "ID Transação", "Date": "Data", "Customer ID": "Matrícula Cliente", "Gender": "Genêro", "Age": "Idade", "Quantity": "Quantidade","Product Category": "Categoria", "Price per Unit": "Preço por unidade", "Total Amount": "Faturamento"})
```

```python
df["Categoria"] = df["Categoria"].replace(to_replace='Beauty', value='Beleza', regex=True)
df["Categoria"] = df["Categoria"].replace(to_replace='Clothing', value='Vestuário', regex=True)
df["Categoria"] = df["Categoria"].replace(to_replace='Electronics', value='Eletrodomésticos', regex=True)
```

```python
df.to_csv('./retail_sales_datasets.csv', index=False)
```

```python
codigo_lojas_df = pd.read_csv('./codigo_lojas.csv')

available_codes = list(codigo_lojas_df['COD_Lojas'])

def generate_random_code():
    return random.choice(available_codes)

sales_df = pd.read_csv('./retail_sales_datasets.csv')

metodos_pagamento = ['Cartão de Crédito', 'Cartão de Débito', 'Boleto', 'Pix', 'Dinheiro']

def gerar_metodo_pagamento_aleatorio():
    return np.random.choice(metodos_pagamento)
    
sales_df['Metodo_Pagamento'] = sales_df.apply(lambda x: gerar_metodo_pagamento_aleatorio(), axis=1)

sales_df['Codigo Loja'] = [generate_random_code() for _ in range(len(sales_df))]

sales_df.to_csv('./retail_sales_datasets.csv', index=False)
```
